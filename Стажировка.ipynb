```python

import spacy
from spacy.matcher import PhraseMatcher

nlp = spacy.load("ru_core_news_sm")

def load_connectors_and_intro_words(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return [line.strip() for line in file if line.strip()]

connectors = load_connectors_and_intro_words('D:\\ДПО Компьютерная Лингвистика\\коннекторы.txt')
intro_words = load_connectors_and_intro_words('D:\\ДПО Компьютерная Лингвистика\\вводные конструкции.txt')

def process_text_file(file_path):
    with open(file_path, 'r', encoding='windows-1251') as file:
        text = file.read().lower()  
    return text

text_file_path = 'D:\\Программирование\\dostoevsky_crimeru\\dostoevsky_crimeru — копия.txt'
text = process_text_file(text_file_path)

doc = nlp(text)

matcher_intro = PhraseMatcher(nlp.vocab)
matcher_connector = PhraseMatcher(nlp.vocab)

patterns_intro = [nlp(text) for text in intro_words]
matcher_intro.add("INTRO", patterns_intro)

patterns_connector = [nlp(text) for text in connectors]
matcher_connector.add("CONNECTOR", patterns_connector)

matches_intro = matcher_intro(doc)
matches_connector = matcher_connector(doc)

found_words = []

for match_id, start, end in matches_intro:
    matched_span = doc[start:end]
    if (start > 0 and doc[start - 1].text == ',') or (end < len(doc) and doc[end].text == ','):
        found_words.append(matched_span.text)
        
for match_id, start, end in matches_connector:
    matched_span = doc[start:end]
    found_words.append(matched_span.text)


print("Найденные вводные конструкции и коннекторы:")
for keyword in found_words:
    print(keyword)
```
